```python
#導入 matplotlib.pyplot為plt
#從bs4 導入 BeautifulSoup
#導入 requests 模組
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline

#宣告month, high, low三個list
month,high,low = [],[],[]
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?response=html&date=20200101&stockNo=2330'

#使用requests get url
html = requests.get(url)
html.encoding = 'UTF-8'

#使用BeautifulSoup解讀
sp = BeautifulSoup(html.text, 'lxml')

#抓出table，並存成變數datas
datas = sp.select("table")[0]
#datas

#在datas中選擇tbody tr的標籤
temp1 = datas.find('tbody')
temp1 = temp1.find_all('tr')

#使用迴圈找出td，讀取出每個月的資料
da = []
for i in temp1:
    da.append(i.find_all('td'))

#td的第一個欄位為年份，不管他
#td的第二個欄位為月份，加入到month list中
#td的第三個欄位為當月最高價，加入到high list中
#td的第四個欄位為當月最低價，加入到low list中
month_list = []
high_list = []
low_list = []

for i in da:
    month_list.append(float(i[1].text))
    high_list.append(float(i[2].text))
    low_list.append(float(i[3].text))
    
#將圖表顯示出來
plt.title('2330')
plt.xlabel('Month')
plt.ylabel('Money')
p1 = plt.plot(month_list, high_list)
p2 = plt.plot(month_list, low_list, color = 'orange')
plt.legend((p1[0], p2[0]), ('Highest', 'Lowest')) 
plt.show()
```
